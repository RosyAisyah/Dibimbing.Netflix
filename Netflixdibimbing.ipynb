```python
for column in data.columns:
    if data[column].dtype == 'object':
        # Jika kolom bertipe objek (teks), isi dengan modus (nilai yang paling sering muncul)
        mode_value = data[column].mode()[0]
        data[column] = data[column].fillna(mode_value)
    else:
        # Jika kolom numerik, isi dengan nilai rata-rata
        mean_value = data[column].mean()
        data[column] = data[column].fillna(mean_value)
```

```python
# Mengecek apakah ada duplicate di seluruh kolom
check_duplicate = data.duplicated().sum()

print(f"Jumlah data yang duplikat = {check_duplicate}")
```

```python

# Handling duplicate
data = data.drop_duplicates()

#tapi disni data saya sudah tidak ada data yang duplikat
```

```python
# Mengecek duplicate setelah di-handle
handle_duplicate = data.duplicated().sum()

print(f"Jumlah data yang duplikat = {handle_duplicate}")
```

```python
# Menampilkan jumlah missing value per kolom
missing_values = data.isnull().sum()
print("Missing value sebelum ditangani:\n")
print(missing_values[missing_values > 0])

# Mengisi missing value dengan "Unknown" untuk kolom teks
data['director'] = data['director'].fillna("Unknown")
data['cast'] = data['cast'].fillna("Unknown")
data['country'] = data['country'].fillna("Unknown")
data['date_added'] = data['date_added'].fillna("Unknown")
data['rating'] = data['rating'].fillna("Unknown")
data['duration'] = data['duration'].fillna("Unknown")
data['description'] = data['description'].fillna("Unknown")

# Cek ulang setelah penanganan
print("\nMissing value setelah ditangani:\n")
print(data.isnull().sum())
```

```python
# Cek jumlah duplikat
print("Jumlah data duplikat:", data.duplicated().sum())

# Hapus duplikat (kalau ada)
data = data.drop_duplicates()

# Cek ulang
print("Jumlah data duplikat setelah dibersihkan:", data.duplicated().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Boxplot release_year
plt.figure(figsize=(10, 4))
sns.boxplot(x=data['release_year'])
plt.title("Distribusi Outlier: release_year")
plt.show()

# Deteksi outlier release_year
q1 = data['release_year'].quantile(0.25)
q3 = data['release_year'].quantile(0.75)
iqr = q3 - q1
batas_bawah = q1 - 1.5 * iqr
batas_atas = q3 + 1.5 * iqr

outliers = data[(data['release_year'] < batas_bawah) | (data['release_year'] > batas_atas)]
print(f"Jumlah outlier pada release_year: {len(outliers)}")
print(outliers[['title', 'release_year']])
```

```python
data = data[(data['release_year'] >= batas_bawah) & (data['release_year'] <= batas_atas)]
```

```python
# Encoding kategorikal
data_encoded = pd.get_dummies(data, columns=['type', 'rating'], drop_first=True)

print("Contoh data setelah encoding:")
print(data_encoded.head())
```

```python
from collections import Counter

# Buat list semua genre
all_genres = data['listed_in'].dropna().str.split(', ')

# Flatten list dan hitung jumlah tiap genre
genre_counter = Counter([genre for sublist in all_genres for genre in sublist])
genre_df = pd.DataFrame(genre_counter.items(), columns=['Genre', 'Jumlah']).sort_values(by='Jumlah', ascending=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.barplot(data=genre_df.head(10), x='Jumlah', y='Genre', palette='viridis')
plt.title('Top 10 Genre Terbanyak di Netflix')
plt.xlabel('Jumlah Tayangan')
plt.ylabel('Genre')
plt.show()
```

```python
# Sama kayak genre, tapi untuk negara
all_countries = data['country'].dropna().str.split(', ')
country_counter = Counter([country for sublist in all_countries for country in sublist])
country_df = pd.DataFrame(country_counter.items(), columns=['Negara', 'Jumlah']).sort_values(by='Jumlah', ascending=False)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=country_df.head(10), x='Jumlah', y='Negara', palette='rocket')
plt.title('Top 10 Negara Asal Tayangan di Netflix')
plt.xlabel('Jumlah Tayangan')
plt.ylabel('Negara')
plt.show()
```
